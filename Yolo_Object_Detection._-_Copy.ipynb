# Yolo Object Detection.

### import some libraries

```python
import cv2
import os
import numpy as np
import time
import glob
import matplotlib.pyplot as plt
```

### load yolo weights and cfg

```python
cfg_path = os.path.join("yolo","yolov3.cfg.txt")
weight_path = os.path.join("yolo","yolov3.weights")
```

### load the neural net in cv2

```python
net = cv2.dnn.readNetFromDarknet(cfg_path,weight_path)
```

### get Layers Names

```python
names =net.getLayerNames()
```

### load the test video

```python
# Importing all necessary libraries
import cv2
import os

# Read the video from specified path
cam = cv2.VideoCapture("lane-only/project_l_h.avi")

try:
	
	# creating a folder named data
	if not os.path.exists('data14'):
		os.makedirs('data14')

# if not created then raise error
except OSError:
	print ('Error: Creating directory of data')

# frame
currentframe = 0

while(True):
	
	# reading from frame
	ret,frame = cam.read()

	if ret:
		# if video is still left continue creating images
		name = './data14/frame' + str(currentframe) + '.jpg'


		# writing the extracted images
		cv2.imwrite(name, frame)

		# increasing counter so that it will
		# show how many frames are created
		currentframe += 1
	else:
		break
        

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```

```python
Frames = []
for i in range(0,currentframe,3):
    img = cv2.imread("data14/frame"+ ""+str(i)+".jpg")
    img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    height, width, layers = img.shape
    size = (width,height)
    Frames.append(img)
```

```python
plt.imshow(Frames[3])
```

```python
(H,W)=Frames[20].shape[:2]
layers_names =[names[i[0]-1] for i in net.getUnconnectedOutLayers()]
layers_names
```

### load labels names

```python
labels_path = os.path.join("yolo","coco.names.txt")
labels=open (labels_path).read().strip().split("\n")
```

### run the interface on the test image

```python
#cal the time of our algo:
start_t =time.time()

for j in range(len(Frames)):
    boxes =[]
    confidences=[]
    classIDs=[]
     
    blob = cv2.dnn.blobFromImage(Frames[j], 1/255.0, (416,416), crop=False, swapRB=False)
    net.setInput(blob)
    layers_output =net.forward(layers_names)
    
    for output in layers_output:
        for dection in output:
            scores = dection[5:]
            classID =np.argmax(scores)
            confidence=scores[classID]

            if confidence> 0.85 :
                box = dection[:4] * np.array([W,H,W,H])
                bx,by,bw,bh= box.astype("int")

                x=int(bx-(bw/2))
                y=int(by-(bh/2))


                boxes.append([x,y,int(bw),int(bh)])
                confidences.append(float(confidence))
                classIDs.append(classID)
                
                
    idxs =cv2.dnn.NMSBoxes(boxes,confidences,0.8,0.8)

    #plot the bounding boxes in the image:
    if len(idxs) > 0:
    
        for i in idxs.flatten():
            (x,y)=[boxes[i][0],boxes[i][1]]
            (w,h)=[boxes[i][2],boxes[i][3]]
            cv2.rectangle(Frames[j],(x,y),(x+w,y+h),(0,255,255),2)
            cv2.putText(Frames[j],"{}"":" "{}".format(labels[classIDs[i]],confidences[i]),(x,y-5),cv2.FONT_HERSHEY_SIMPLEX\
                        ,0.5,(0,139,139),2)


    

    
print("it's take about {} min".format((time.time()-start_t)/60))
```
